```python
import pandas as pd
import matplotlib.pyplot as plt 
from matplotlib.ticker import FormatStrFormatter

EPOCHS = 10
comms_round = 20
```

```python
# FedAvg files
global_stats_mae_FedAvg = pd.read_csv('global_stats_mae_FedAvg.csv')
global_stats_rmse_FedAvg = pd.read_csv('global_stats_rmse_FedAvg.csv')
global_stats_r2_FedAvg = pd.read_csv('global_stats_r2_FedAvg.csv')

# FedProx files
global_stats_mae_FedProx = pd.read_csv('global_stats_mae_FedProx.csv')
global_stats_rmse_FedProx = pd.read_csv('global_stats_rmse_FedProx.csv')
global_stats_r2_FedProx = pd.read_csv('global_stats_r2_FedProx.csv')

# KRUM files
global_stats_mae_KRUM = pd.read_csv('global_stats_mae_KRUM.csv')
global_stats_rmse_KRUM = pd.read_csv('global_stats_rmse_KRUM.csv')
global_stats_r2_KRUM = pd.read_csv('global_stats_r2_KRUM.csv')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(30, 7))

# Plot MAE for FedAvg
ax1.plot(global_stats_mae_FedAvg[global_stats_mae_FedAvg.columns[1:11]], label='train')
ax1.set_ylabel('MAE', fontsize=25)
ax1.set_xlabel('Round', fontsize=25)
ax1.tick_params(axis='both', which='major', labelsize=20)
ax1.xaxis.set_major_formatter(FormatStrFormatter('%.0f'))
ax1.text(0.5, 0.95, 'FedAvg', fontsize=25, ha='center', va='center', transform=ax1.transAxes,
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))

# Plot MAE for FedProx
ax2.plot(global_stats_mae_FedProx[global_stats_mae_FedProx.columns[1:11]], label='train')
ax2.set_ylabel('MAE', fontsize=25)
ax2.set_xlabel('Round', fontsize=25)
ax2.tick_params(axis='both', which='major', labelsize=20)
ax2.xaxis.set_major_formatter(FormatStrFormatter('%.0f'))
ax2.text(0.5, 0.95, 'FedProx', fontsize=25, ha='center', va='center', transform=ax2.transAxes,
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))

# Plot MAE for KRUM
ax3.plot(global_stats_mae_KRUM[global_stats_mae_KRUM.columns[1:11]], label='train')
ax3.set_ylabel('MAE', fontsize=25)
ax3.set_xlabel('Round', fontsize=25)
ax3.tick_params(axis='both', which='major', labelsize=20)
ax3.xaxis.set_major_formatter(FormatStrFormatter('%.0f'))
ax3.text(0.5, 0.95, 'KRUM', fontsize=25, ha='center', va='center', transform=ax3.transAxes,
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))

plt.savefig(f'MAE_edge_{EPOCHS}_{comms_round}.png', bbox_inches='tight', dpi=400)
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(30, 7))
plt.rcParams.update({'font.size': 25})

# Plot MAE
ax1.plot(global_stats_mae_FedAvg['A'], label='train', marker='o')
ax1.set_ylabel('MAE', fontsize=25)
ax1.set_xlabel('Round', fontsize=25)
ax1.tick_params(axis='both', which='major', labelsize=20)
ax1.xaxis.set_major_formatter(FormatStrFormatter('%.0f'))
ax1.text(0.5, 0.95, 'FedAvg', fontsize=25, ha='center', va='center', transform=ax1.transAxes,
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))

# Plot MAE
ax2.plot(global_stats_mae_FedProx['A'], label='train', marker='o')
ax2.set_ylabel('MAE', fontsize=25)
ax2.set_xlabel('Round', fontsize=25)
ax2.tick_params(axis='both', which='major', labelsize=20)
ax2.xaxis.set_major_formatter(FormatStrFormatter('%.0f'))
ax2.text(0.5, 0.95, 'FedProx', fontsize=25, ha='center', va='center', transform=ax2.transAxes,
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))

# Plot MAE
ax3.plot(global_stats_mae_KRUM['A'], label='train', marker='o')
ax3.set_ylabel('MAE', fontsize=25)
ax3.set_xlabel('Round', fontsize=25)
ax3.tick_params(axis='both', which='major', labelsize=20)
ax3.xaxis.set_major_formatter(FormatStrFormatter('%.0f'))
ax3.text(0.5, 0.95, 'KRUM', fontsize=25, ha='center', va='center', transform=ax3.transAxes,
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))

# Add text annotations for the last round values
round_idx = len(global_stats_mae_FedAvg) - 1
ax1.text(round_idx, global_stats_mae_FedAvg['A'][round_idx], f'{global_stats_mae_FedAvg["A"][round_idx]:.2f}', fontsize=20, bbox=dict(facecolor='white', alpha=0.5, edgecolor='black'))
ax2.text(round_idx, global_stats_mae_FedProx['A'][round_idx], f'{global_stats_mae_FedProx["A"][round_idx]:.2f}', fontsize=20, bbox=dict(facecolor='white', alpha=0.5, edgecolor='black'))
ax3.text(round_idx, global_stats_mae_KRUM['A'][round_idx], f'{global_stats_mae_KRUM["A"][round_idx]:.2f}', fontsize=20, bbox=dict(facecolor='white', alpha=0.5, edgecolor='black'))

plt.savefig(f'MAE_global_{EPOCHS}_{comms_round}.png', bbox_inches='tight', dpi=400)
plt.tight_layout()
plt.show()
```
